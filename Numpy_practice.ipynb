Can access values using python index notation: x1[0] or for multidimensional separate dimensions with commas i.e. x2[0,0]
Likewise we can access subarray by using the samy python sintax x2[::-1, ::-1] -> reverses both rows and columns. x2[:2, :2] will show 0,1 row and column.

IMPORTANT DETAIL: Unlike in python lists, when we slice using NumPy arrays we are not getting a copy but rather a view. That means that whatever change made to the sliced array will also take place to the original array.

```python
print(x2)
```

```python
x2_sub = x2[:2, :2]
print(x2_sub)
```

```python
x2_sub[0,0] = 99
print(x2_sub)
print(x2)
```

Note how even though the change occured in x2_sub, the x2 array changed as well. To make a different space in memory use the copy() method.

```python
x2_sub_copy = x2[:2,:2].copy()
print(x2_sub_copy)
```

```python
x2_sub_copy[0,0] = 10
print(x2_sub_copy)
print(x2)
```

Note how the original array did not change
